```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
supplier_data1 = pd.read_excel("supplier_data1.xlsx")
supplier_data2 = pd.read_excel("supplier_data2.xlsx")
buyer_preferences = pd.read_excel("buyer_preferences.xlsx")
```

```python
import numpy as np


supplier_data2["Finish"] = np.where(
    supplier_data2["Description"] == "Material is Oiled", "gebeizt",
    np.where(
        supplier_data2["Description"] == "Material is Painted", "gebeizt und geglüht",
        np.where(
            supplier_data2["Description"] == "Material is not Oiled", "ungebeizt",
            "Unknown"  
        )
    )
)
```

```python
supplier_data_finish_left = supplier_data1.merge(supplier_data2, how = 'left', on = "Finish")
```

```python
supplier_data_finish_left
```

```python
supplier_data_finish_left = supplier_data_finish_left[~(supplier_data_finish_left["Gross weight (kg)"] < supplier_data_finish_left["Weight (kg)"])]
```

```python
supplier_data_finish_left.reset_index(drop=True, inplace=True)
supplier_data_finish_left
```

```python
buyer_preferences
```

```python
supplier_data_finish_left['Quantity'] = supplier_data_finish_left['Quantity_y']
supplier_data_finish_left = supplier_data_finish_left.drop(columns = 'Quantity_y')
```
